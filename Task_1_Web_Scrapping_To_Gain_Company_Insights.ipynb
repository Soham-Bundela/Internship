```python
# All the parameter tags are converted into Meaningful Information

def cleanedParameterTable(review):
  review_parameter =[]
  for i in review:
    a = i.find_all('td', class_='review-rating-header')
    review_parameter.append(a)

  # Remove "<td class="review-rating-header" and '</td>' prefix and double quotes from comments
  cleaned_parameter = []
  for rev in review_parameter:
    for tag in rev:
      # Get the text content of the tag
      tag_text = tag.get_text()
      # To remove the tags
      cleaned_text = re.sub(r'["]', '', tag_text)
      cleaned_parameter.append(cleaned_text)

  return cleaned_parameter
```

```python
# Function to get name of the Passangers

def getNames(review_list): # Renamed the input parameter for clarity
  # To extract the names of the passanger
  names = []
  for review in review_list: # Iterate through the list of review bodies
      # Find the span tag with itemprop="name"
      name_tag = review.find('span', itemprop="name") # Use find instead of find_all as we expect one name
      if name_tag:
          # Get the text content of the tag and clean it
          name = name_tag.get_text(strip=True)
          # Remove square brackets if they are present in the text (from previous cleaning attempts)
          cleaned_name = re.sub(r'[\[\]]', '', name)
          names.append(cleaned_name)
      else:
          # Append None if the name tag is not found for this review
          names.append(None)

  return names
```

```python
# Previewing getNames Function
getNames(review)
```

```python
# Function to get name of the country of residence of the passenger

def getCountryName(review_list):
  country_names = []
  for review in review_list: # Iterate through the list containing typically one review body
    # Find the h3 tag containing the name and country
    h3_tag = review.find('h3', class_="text_sub_header userStatusWrapper")
    if h3_tag:
      tag_text = h3_tag.get_text()
      # Use regex to find the text within parentheses (the country name)
      match = re.search(r'\((.*?)\)', tag_text)
      if match:
        country = match.group(1).strip() # strip() to remove potential whitespace
        country_names.append(country)
      else:
        country_names.append(None) # Append None if no country found in parentheses
    else:
      country_names.append(None) # Append None if the h3 tag is not found

  return country_names
```

```python
# Function to get Model of the flight

def getAircraftModel(review):
  aircraft_models = []
  for i in review:
      # Find the 'Aircraft' header tag
      aircraft_header_tag = i.find('td', class_="review-rating-header aircraft")
      if aircraft_header_tag:
          # Find the next sibling 'td' tag with class 'review-value'
          aircraft_value_tag = aircraft_header_tag.find_next_sibling('td', class_='review-value')
          if aircraft_value_tag:
              # Extract the text content
              model_name = aircraft_value_tag.get_text(strip=True)
              aircraft_models.append(model_name)
          else:
              # Append None if the value tag is not found
              aircraft_models.append(None)
      else:
          # Append None if the header tag is not found
          aircraft_models.append(None)

  return aircraft_models
```

```python
# Is the flight Recommanded by the Passenger to others

def getRecommandation(review):
  recommended_status = []
  for i in review:
      # Find the 'Recommended' header tag
      recommended_header_tag = i.find('td', class_="review-rating-header recommended")
      if recommended_header_tag:
          # Find the next sibling 'td' tag with class 'review-value'
          recommended_value_tag = recommended_header_tag.find_next_sibling('td', class_='review-value')
          if recommended_value_tag:
              # Extract the text content (which will be 'yes' or 'no')
              status = recommended_value_tag.get_text(strip=True)
              recommended_status.append(status)
          else:
              recommended_status.append(None) # Append None if the value tag is not found
      else:
          recommended_status.append(None) # Append None if the header tag is not found

  return recommended_status
```

```python
# To get the Review Text by the Passenger/Commentator

def getReviewText(soup):
  # Storing all the Written Review in an array called "comment"
  comment =[]
  for i in soup.find_all('div', class_='text_content', itemprop='reviewBody'):
    comment.append(i.get_text())

  # Remove "✅ Trip Verified | " prefix and double quotes from comments
  cleaned_comments = []
  for review in comment:
    # Remove "✅ Trip Verified | "
    cleaned_review = re.sub(r'✅ Trip Verified \|?\s*', '', review)
    # Remove double quotes
    cleaned_review = re.sub(r'"', '', cleaned_review)
    cleaned_comments.append(cleaned_review)
  return cleaned_comments
```

# Including all the WebPages into the DataFrame

```python
# Function to extract data from all the 397 pages in the website

def scrape_airline_reviews(base_url, num_pages):
    all_reviews_data = {
        'Name': [],
        'Country': [],
        'Date Flown': [],
        'Aircraft Model': [],
        'Type Of Traveller': [],
        'Cabin Type': [],
        'Route': [],
        'Seat Comfort Rating': [],
        'Cabin Staff Service Rating': [],
        'Food & Beverages Rating': [],
        'Inflight Entertainment Rating': [],
        'Ground Service Rating': [],
        'Wifi & Connectivity Rating': [],
        'Value For Money Rating': [],
        'Recommended': [],
        'Review Text': []
    }

    for page_num in range(1, num_pages + 1):
        url = f"{base_url}/page/{page_num}/"
        print(f"Scraping page: {url}") # Added print statement to show progress

        try:
            response = requests.get(url)
            response.raise_for_status()  # Raise an exception for bad status codes
            soup = BeautifulSoup(response.text, "html.parser")

            # Find all review bodies on the current page
            reviews_on_page = soup.find_all('div', class_='body')

            if not reviews_on_page:
                print(f"No reviews found on page {page_num}. Stopping.")
                break # Stop if no reviews are found on a page

            # Extract data using the previously defined functions for the current page's reviews
            # Ensure each function handles missing data and returns a list of length len(reviews_on_page)
            all_reviews_data['Name'].extend(getNames(reviews_on_page))
            all_reviews_data['Country'].extend(getCountryName(reviews_on_page))
            all_reviews_data['Date Flown'].extend(getDateFlown(reviews_on_page))
            all_reviews_data['Aircraft Model'].extend(getAircraftModel(reviews_on_page))
            all_reviews_data['Type Of Traveller'].extend(getTravellerType(reviews_on_page))
            all_reviews_data['Cabin Type'].extend(getCabinType(reviews_on_page))
            all_reviews_data['Route'].extend(getRoute(reviews_on_page))
            all_reviews_data['Seat Comfort Rating'].extend(getSeatComfort(reviews_on_page))
            all_reviews_data['Cabin Staff Service Rating'].extend(getCabinStaffService(reviews_on_page))
            all_reviews_data['Food & Beverages Rating'].extend(getFoodAndBeverageService(reviews_on_page))
            all_reviews_data['Inflight Entertainment Rating'].extend(getInflightEntertainmentService(reviews_on_page))
            # Using the corrected getGroundService function
            all_reviews_data['Ground Service Rating'].extend(getGroundService(reviews_on_page))
            all_reviews_data['Wifi & Connectivity Rating'].extend(getWifiAndConnectivity(reviews_on_page))
            all_reviews_data['Value For Money Rating'].extend(getValueForMoney(reviews_on_page))
            all_reviews_data['Recommended'].extend(getRecommandation(reviews_on_page))

            # Extract review text for the current page
            review_texts_on_page = []
            for review_body in reviews_on_page:
                 text_content_tag = review_body.find('div', class_='text_content', itemprop='reviewBody')
                 if text_content_tag:
                     # Clean the text content by removing the prefix and stripping whitespace
                     cleaned_review = re.sub(r'✅ Trip Verified \|?\s*', '', text_content_tag.get_text(strip=True))
                     review_texts_on_page.append(cleaned_review)
                 else:
                     review_texts_on_page.append(None) # Append None if review text not found

            all_reviews_data['Review Text'].extend(review_texts_on_page)


        except requests.exceptions.RequestException as e:
            print(f"Error fetching page {url}: {e}")
            continue # Continue to the next page if there's an error
        except Exception as e:
            print(f"An unexpected error occurred while processing page {url}: {e}")
            continue # Continue to the next page for other errors as well


    # Create a Pandas DataFrame from the collected data
    df_all_reviews = pd.DataFrame(all_reviews_data)

    return df_all_reviews
```

```python
# Entering the base url of the website
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"

# No. of pages to scrap
num_pages_to_scrape = 395

# Using the function
df_all_reviews = scrape_airline_reviews(base_url, num_pages_to_scrape)
```

```python
# Split the 'Route' column into 'Departed_From' and 'Arrived_At'
route_split = df['Route'].str.split(' to ', expand=True)

# Handle routes with 'via'
route_split_to = route_split[1].str.split(' via ', expand=True)

# Create 'Departed_From' and 'Arrived_At' columns
df['Departed_From'] = route_split[0]
df['Arrived_At'] = route_split_to[0]

# Extract the 'Via' information
df['Via'] = route_split_to[1]

# Update 'Arrived_At' for direct flights (where there was no 'via')
df['Arrived_At'] = df['Arrived_At'].fillna(route_split[1])


# Drop the original 'Route' column
df.drop('Route', axis=1, inplace=True)
```

```python
# Airplanes operated by British Airways are :
# Airbus A319, A320, A321, A350-1000, A380, and Boeing 777 and 787 family aircraft, along with Embraer 190s

def classify_aircraft(model):
    if pd.isna(model):
        return 'Unknown'
    model = model.lower()
    if 'airbus a319' in model or '319' in model:
        return 'Airbus A319'
    elif 'airbus a320' in model or '320' in model:
        return 'Airbus A320'
    elif 'airbus a321' in model or '321' in model:
        return 'Airbus A321'
    elif 'airbus a350-1000' in model or '350' in model:
        return 'Airbus A350-1000'
    elif 'airbus a380' in model or '380' in model:
        return 'Airbus A380'
    elif 'boeing 777' in model or '777' in model:
        return 'Boeing 777 Family'
    elif 'boeing 787' in model or '787' in model:
        return 'Boeing 787 Family'
    elif 'embraer 190' in model or '190' in model:
        return 'Embraer 190'
    else:
        return 'Other'

df['Aircraft_Category'] = df['Aircraft Model'].apply(classify_aircraft)
df.drop('Aircraft Model', axis=1, inplace=True)
display(df['Aircraft_Category'].value_counts())
```

```python
# Define the mapping for Aircraft_Category
aircraft_category_mapping = {
    'Unknown': 0,
    'Other': 1,
    'Airbus A319': 2,
    'Airbus A320': 3,
    'Airbus A321': 4,
    'Airbus A350-1000': 5,
    'Airbus A380': 6,
    'Boeing 777 Family': 7,
    'Boeing 787 Family': 8,
    'Embraer 190': 9
}

# Apply the mapping to create a new numerical column
df['Aircraft_Category_Encoded'] = df['Aircraft_Category'].map(aircraft_category_mapping)

# Fill any potential NaN values with 0 (though all categories should be mapped)
df['Aircraft_Category_Encoded'] = df['Aircraft_Category_Encoded'].fillna(0).astype(int)
```
